```python
import json
import pandas as pd

# Cargar el JSON
with open('../src/clases_combinadas.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Crear DataFrame
df = pd.DataFrame(data)

# Añadir columnas para nombres en español e inglés (vacías por defecto)
df['planta_español'] = ""
df['planta_ingles'] = df['planta']
df['nombre_comun_español'] = ""
df['nombre_comun_ingles'] = df['nombre_comun']
# nombre_cientifico y clasificacion ya existen

# Reordenar columnas
cols = [
    'planta_español', 'planta_ingles',
    'nombre_comun_español', 'nombre_comun_ingles',
    'nombre_cientifico', 'clasificacion'
]
df_final = df[cols]
```

```python
# Modificar df_final según las condiciones indicadas
mask_vid = df_final['planta_ingles'] == 'Vid'

# Guardar los valores originales para invertirlos
nombre_comun_español_orig = df_final.loc[mask_vid, 'nombre_comun_español'].copy()
nombre_comun_ingles_orig = df_final.loc[mask_vid, 'nombre_comun_ingles'].copy()

# Realizar los cambios
df_final.loc[mask_vid, 'planta_español'] = 'Vid'
df_final.loc[mask_vid, 'planta_ingles'] = ''
df_final.loc[mask_vid, 'nombre_comun_español'] = nombre_comun_ingles_orig
df_final.loc[mask_vid, 'nombre_comun_ingles'] = nombre_comun_español_orig

```

```python
df_final.to_excel('../src/clases_a_rellenar.xlsx', index=False)

print("✅ Archivo Excel generado en src/clases_a_rellenar.xlsx")
df_final.head()
```
